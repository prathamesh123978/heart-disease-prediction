# 01 — Exploratory Data Analysis
Heart Disease Prediction Project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
df = pd.read_csv('../data/heart.csv')
print('Shape:', df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print('Missing values:')
print(df.isnull().sum())
```

```python
# Target distribution
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df, palette='Set2')
plt.title('Target Distribution (0=No Disease, 1=Disease)')
plt.savefig('../report/target_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Age distribution by target
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='age', hue='target', kde=True, palette='Set2')
plt.title('Age Distribution by Heart Disease')
plt.savefig('../report/age_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.savefig('../report/correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Boxplots for numerical features
num_cols = ['age','trestbps','chol','thalach','oldpeak']
fig, axes = plt.subplots(1, len(num_cols), figsize=(18,4))
for ax, col in zip(axes, num_cols):
    sns.boxplot(x='target', y=col, data=df, palette='Set2', ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.savefig('../report/boxplots.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Categorical feature counts
cat_cols = ['sex','cp','fbs','restecg','exang','slope','ca','thal']
fig, axes = plt.subplots(2, 4, figsize=(18,8))
for ax, col in zip(axes.flatten(), cat_cols):
    sns.countplot(x=col, hue='target', data=df, palette='Set2', ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.savefig('../report/categorical_features.png', dpi=150, bbox_inches='tight')
plt.show()
```
